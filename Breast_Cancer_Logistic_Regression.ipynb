## Importing Tools

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import timeit
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
```

## Fetching the dataset

```python
df = pd.read_csv('data.csv')
```

## Quick Look at the dataset

```python
df.info()
```

##### There are 569 instances in the dataset. All attributes are numerical, except the "diagnosis". It's type is object

```python
df["diagnosis"].value_counts()
```

##### Converting "diagnosis" field into a numerical column

```python
df["diagnosis"] = df["diagnosis"].map({'M':1 ,'B':0})
```

```python
df["diagnosis"].value_counts()
```

##### "Unnamed: 32" field has 0 non-null values. So droping this field.

```python
#data.replace(np.nan, 0, inplace=True)
df = df.drop(["Unnamed: 32", "id"], axis=1)
```

```python
df.describe()
```

```python
df.hist(bins=50, figsize=(20,15))
plt.show()
```

######  Correlation of features with the diagnosis column

```python
corr_matrix = df.corr(numeric_only = True).abs()
corr_matrix["diagnosis"].sort_values(ascending=False)
```

######  Taking features that have corelation value more than 0.1 with the diagnosis feature

```python
df = df[corr_matrix["diagnosis"][corr_matrix["diagnosis"]>0.1].index]
```

###### Spliting the data into train (80%) and test set (20%)

```python
X = df.drop(["diagnosis"], axis=1)
y = df["diagnosis"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42)
```

###### Scaling the features 

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X_train)
X_test_minmax = min_max_scaler.transform(X_test)
```

###### Classification using Logistic Regresssion

```python
clf = LogisticRegressionCV(random_state=42, max_iter=500).fit(X_train_minmax, y_train)
clf.score(X_train_minmax, y_train)
```

```python
clf.score(X_test_minmax, y_test)
```
